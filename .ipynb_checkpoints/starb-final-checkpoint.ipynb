# Starbucks Capstone Challenge

## Project overview

This data set contains simulated data that mimics customer behavior on the Starbucks rewards mobile app. 
Starbucks sends its users an offer every few days. It is either an informational offer, with a discount or a buy one get one free(BOGO) type. The data does not show what products were purchased, only information about the cost of each order.

### Problem to solve
Using the data provided, we will try to answer some questions: 

 - Which users would buy the product despite not getting offers.
 - Whether we can predict whether a given user will respond to a given
   offer.
 - Which demographic groups respond best to which offer type.
 
### Data sets
The data is contained in three files.
Brief description of the data and specific features:

 #### 1. portfolio.json - containing offer ids and meta data about each offer:
 - id (string) - offer id
 - offer_type (string) - type of offer ie BOGO, discount, informational
 - difficulty (int) - minimum required spend to complete an offer
 - reward (int) - reward given for completing an offer
 - duration (int) - time for offer to be open, in days
 - channels (list of strings)
 
#### 2. profile.json - demographic data for each customer:

 - age (int) - age of the customer
 - became_member_on (int) - date when customer created an app account
 - gender (str) - gender of the customer (note some entries contain 'O' for other rather than M or F)
 - id (str) - customer id
 - income (float) - customer's income
 
#### 3.transcript.json - records for transactions, offers received, offers viewed, and offers completed:

 - event (str) - record description (ie transaction, offer received,
   offer viewed, etc.)
 - person (str) - customer id
 - time (int) - time in hours since start of test. The data begins at
   time t=0
 - value - (dict of strings) - either an offer id or transaction amount
   depending on the record

There is no NaN values, so now we need to get the *value* feature in order. It contains dictionaries with various variables. Let's extract them into separate columns. 
It seems problematic to extract the *offer id* from two different event types, as they have different keys *'offer id'* and *'offer_id'*. The easiest way is to write a function to pull this variable.

```python
def get_offer_id(element):
    '''
    extracts offer id from the library, regardless of whether the key has a "_" or not
    
    INPUT:
    element : element from transcript value feature
    
    OUTPUT:
    offer_id : offer id from offer received or offer completed
    
    '''
    if element.get('offer id') is None:
        offer_id = element.get('offer_id')
    else:
        offer_id = element.get('offer id')
    return offer_id
```

```python
#getting different features from value

transcript['offer id']= transcript['value'].apply(lambda x: get_offer_id(x))
transcript['reward'] = transcript['value'].apply(lambda x: x.get('reward'))
transcript['amount'] = transcript['value'].apply(lambda x: x.get('amount'))
transcript.drop(columns=['value'],inplace=True)
transcript
```

### Data Preprocessing

After cleaning and analyzing the datasets provided, it is time to merge them into one set and add some additional features such as whether the transaction was using a promotion and whether the user who used the promotion saw it or used it because it was simply available.

First we will combine the transcript and portfolio data which will allow us in the next steps to label whether the offer was seen before it was used.

```python
portfolio.rename(columns={'id':'offer id','reward':'offer reward'},inplace=True)

df_combine = transcript.merge(portfolio,on='offer id',how ='left')
df_combine.head(3)
```

Now we will create two auxiliary dataframes which we will then combine into a third one, which will be the list of offers that the user saw before using. 

```python
#we create a list of offers seen
offer_viewed = df_combine[['person','offer id','duration','time']][df_combine['event']=='offer viewed']
offer_viewed.rename(columns={'time':'time_of_view'},inplace=True)

#we create a list of offers completed

offer_completed = df_combine[['person','offer id','time']][df_combine['event']=='offer completed']
offer_completed.rename(columns={'time':'time_of_completion'},inplace=True)


#Now we combine the two lists above
offer_viewed_completed = offer_completed.merge(offer_viewed, on=['person','offer id'],how='left')

#We calculate the time to use the offer by subtracting the time to see from the time to use it
offer_viewed_completed['time_to_complete'] = offer_viewed_completed['time_of_completion']-offer_viewed_completed['time_of_view']


#Now we remove from this list the offers:
#    -that were seen after they were completed,
#    -that were completed after the offer period was longer than the offer period.
offer_viewed_completed = offer_viewed_completed[(offer_viewed_completed['time_to_complete']>=0)\
                                             &(offer_viewed_completed['time_to_complete']<=offer_viewed_completed['duration'])]


#to avoid assigning offer viewed event to offer completion twice we need to remove duplicates

offer_viewed_completed.drop_duplicates(['person','offer id','time_of_view'], keep='first',inplace=True)
offer_viewed_completed.drop_duplicates(['person','offer id','time_of_completion'], keep='first',inplace=True)

offer_viewed_completed.head(3)
```

```python
#adding info about completed offer to event 'offer completed' in df_combine

offer_viewed_completed.rename(columns={'time_of_view':'time'},inplace=True)
offer_viewed_completed['event']='offer viewed'

df_combine=df_combine.merge(offer_viewed_completed[['person','offer id','time','event','time_to_complete']],\
                       on=['person','offer id','time','event'],how='left')

df_combine.rename(columns={'time_to_complete':'viewed_complated'},inplace=True)

#assigning 1 if offer was viewed and complated and 0 otherwise
df_combine['viewed_complated'].loc[~df_combine['viewed_complated'].isnull()] = 1
df_combine['viewed_complated'] = df_combine['viewed_complated'].fillna(0)

offer_viewed_completed.rename(columns={'time':'time_of_completion'},inplace=True)

df_combine.head(3)
```

```python
df_combine['viewed_complated'].sum()
```

Now we make another supporting dataframe to create a list of transactions that used the promotion.

```python

transactions = transcript[['person','time']][transcript['event']=='transaction']

# getting info about which offer(if so) was used in transaction
offer_completed.rename(columns={'time_of_completion':'time'},inplace=True)
transaction_with_offer = transactions.merge(offer_completed,on=['person','time'],how='left')
transaction_with_offer['event']='transaction'

#We don't need information about what specific bid was used in the transaction, just is it was(1) or not(0).
transaction_with_offer['offer id'].loc[~transaction_with_offer['offer id'].isnull()] = 1
transaction_with_offer.rename(columns={'offer id':'trans with offer'},inplace=True)
transaction_with_offer.drop_duplicates(inplace=True)

transaction_with_offer.head(3)
```

```python
#adding info if transaction was with offer to df_combine
df_combine = df_combine.merge(transaction_with_offer,on=['event','person','time'],how='left')
df_combine['trans with offer'] = df_combine['trans with offer'].fillna(0)
df_combine.head(4)
```

```python
#Finally, we add information about our customers.
profile.rename(columns={'id':'person'},inplace=True) 
df_combine = df_combine.merge(profile,on=['person'],how='left')

#We remove those rows where customer information is not present.
df_combine.dropna(subset=['income'] ,inplace=True)
df_combine.head(5)
```

Finally, we have one dataframe with all the data we need for our models.

### Data Modeling

The first problem we want to solve is to see if, having data about customers and the type of offer, we can resolve whether a given customer, having seen the offer, will use it. If the model achieves satisfactory results we will be able to stop sending offers to people who are not interested in them.
For this type of task, the best seems to be the logistic regression model.


To begin with, we will choose the features that will seem useful here. Our task will be to predict whether the customer will use it after seeing the promotion. So the feature we will want to predict is 'viewed_complied', while information about the type of offer and our customer will be used to fit our model.

```python
df = df_combine.iloc[: , 15:25]

df.head(4)
```

Let's start with just demographic information about our customer.

```python
df.drop_duplicates(inplace=True)

columns_to_drop = ['income_label','age_label','became_member_on','trans with offer']

df.drop(columns=columns_to_drop,inplace=True)
df=df.fillna(0)
df.shape
```

After preparing our data, we can do a split between training and testing data, then fit our model and evaluate it with accuracy metrics.

```python
column='viewed_complated'

X,y=df.drop(columns=[column]),df[column]
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)

logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)

predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
print(score)
```

The accuracy of this model is 76%, for a start it is not bad, but let's try to add one more characteristic of the customer that we did not include before, which is the date of becoming a member.

```python
df = df_combine.iloc[: , 15:25]
df.drop_duplicates(inplace=True)

columns_to_drop = ['income_label','age_label']

df.drop(columns=columns_to_drop,inplace=True)

df=df.fillna(0)

X,y=df.drop(columns=[column]),df[column]

x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=42)

logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)

predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
print(score)
```

After adding the new feature, the accuracy of the model increased to 80%, which is already a good result. Now with the help of Grid Search we will try hyperparameter tuning our model.

```python
# setting some hyperparameters
solvers = ['newton-cg', 'lbfgs']
penalty = ['l2','l1']
c_values = [1000,100, 10, 1.0]

# defining grid search
grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=logisticRegr, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
grid_result = grid_search.fit(X, y)

# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

For this model, all hyperparameters gave similar results, so we can stay with the defaults.
